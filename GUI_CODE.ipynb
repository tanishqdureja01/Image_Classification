```python
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
import numpy as np
import warnings
import os

# Suppress annoying background warnings
warnings.filterwarnings("ignore")
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from keras.models import load_model

# Load the trained model
model = load_model('model1_cifar_100epoch.h5')

# CIFAR-10 dataset classes
classes = { 
    0: 'AEROPLANE',
    1: 'AUTOMOBILE',
    2: 'BIRD',
    3: 'CAT',
    4: 'DEER',
    5: 'DOG',
    6: 'FROG',
    7: 'HORSE',
    8: 'SHIP',
    9: 'TRUCK' 
}

# --- ACCESSIBLE COLOR PALETTE ---
BG_COLOR = '#121212'         # Black background
TEXT_MAIN = '#FFFFFF'        # Pure white
TEXT_ACCENT = '#00E676'      # Neon green title
TEXT_OUTPUT = '#FFEA00'      # Yellow for classification results
BTN_BG = '#1F1F1F'           # Dark gray buttons
BTN_BORDER = '#FFFFFF'       # White border for button

# Initialize GUI
top = tk.Tk()
top.geometry('800x700')
top.title('Accessible Image Classification CIFAR10')
top.configure(background=BG_COLOR)

# Output Labels
label = Label(top, background=BG_COLOR, font=('Arial', 32, 'bold'))
sign_image = Label(top, background=BG_COLOR)

# Global placeholder for the classify button so we can manage it easily
classify_b = None

def classify(file_path):
    try:
        # Open and resize the image
        image = Image.open(file_path)
        image = image.resize((32, 32))
        
        # Convert to numpy array and normalize values to [0, 1]
        image = np.array(image) / 255.0
        
        # Add the batch dimension: shape becomes (1, 32, 32, 3)
        image = np.expand_dims(image, axis=0)
        
        # Predict probabilities and grab the highest scoring class index
        predictions = model.predict(image)
        pred = np.argmax(predictions, axis=1)[0]
        sign = classes[pred]
        
        # Print to terminal for debugging and update GUI output
        print(f"Result: {sign}")
        label.configure(foreground=TEXT_OUTPUT, text=f"RESULT: {sign}") 
    except Exception as e:
        print(f"Error during classification: {e}")
        label.configure(foreground='#FF1744', text="Error classifying image")

def show_classify_button(file_path):
    global classify_b
    # Clean up existing button if it's already there from a previous image
    if classify_b:
        classify_b.destroy()
        
    classify_b = Button(
        top,
        text="[ CLASSIFY IMAGE ]",
        command=lambda: classify(file_path),
        padx=20,
        pady=10,
        bd=2,
        relief="solid"
    )
    classify_b.configure(
        background=BTN_BG, 
        foreground=TEXT_ACCENT,
        activebackground=TEXT_ACCENT,
        activeforeground=BG_COLOR,
        font=('Arial', 16, 'bold')
    )
    classify_b.pack(side=BOTTOM, pady=20)

def upload_image():
    try:
        file_path = filedialog.askopenfilename()
        if not file_path:  # User cancelled the file dialog
            return
            
        uploaded = Image.open(file_path)
        
        # Dynamically scale UI preview to fit comfortably
        uploaded.thumbnail(((top.winfo_width() / 2), (top.winfo_height() / 2)))
        im = ImageTk.PhotoImage(uploaded)
        
        sign_image.configure(image=im)
        sign_image.image = im
        label.configure(text='')  # Clear previous text
        
        show_classify_button(file_path)
    except Exception as e:
        print(f"Error uploading image: {e}")

# High-visibility Heading
heading = Label(
    top, 
    text="IMAGE CLASSIFICATION (CIFAR-10)", 
    pady=20, 
    font=('Arial', 24, 'bold')
)
heading.configure(foreground=TEXT_ACCENT, background=BG_COLOR)
heading.pack(side=TOP, fill=X)

# Main Upload Button Layout
upload = Button(
    top, 
    text="UPLOAD AN IMAGE", 
    command=upload_image,
    padx=20, 
    pady=10,
    bd=2,
    relief="solid"
)
upload.configure(
    background=BTN_BG, 
    foreground=TEXT_MAIN,
    activebackground=TEXT_MAIN,
    activeforeground=BG_COLOR,
    font=('Arial', 16, 'bold')
)

# Pack layout carefully to prevent overlapping
upload.pack(side=BOTTOM, pady=30)
label.pack(side=BOTTOM, pady=15)
sign_image.pack(side=BOTTOM, expand=True, fill=BOTH, pady=10)

top.mainloop()
```
